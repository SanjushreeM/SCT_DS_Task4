```python
zip_path = '/content/Road Traffic Accidents.zip'
extract_path = '/content/accident_data'

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

print("Unzipping completed!")
```

```python
file_path = '/content/accident_data/RTA Dataset.csv'
df = pd.read_csv(file_path)

df.head()
```

```python
plt.figure()
df['Weather_conditions'].value_counts().plot(kind='bar')
plt.title('Accidents by Weather Conditions')
plt.xlabel('Weather Conditions')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure()
df['Road_surface_conditions'].value_counts().plot(kind='bar')
plt.title('Accidents by Road Surface Conditions')
plt.xlabel('Road Surface Conditions')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df['Hour'] = pd.to_datetime(df['Time'], errors='coerce').dt.hour
plt.figure()
plt.hist(df['Hour'].dropna())
plt.title('Accidents by Time of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Accidents')
plt.tight_layout()
plt.show()
```

```python
plt.figure()
df['Area_accident_occured'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Accident Hotspot Areas')
plt.xlabel('Area')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
